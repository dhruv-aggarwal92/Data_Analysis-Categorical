# Total Apps in a single Categorie

```python
c=0
for i in data['Category']:
    if(i == "ART_AND_DESIGN"):
        c += 1
print("There are total",c,"application in ART_AND_DESIGN")
```

# Total Apps in a every Categorie

```python
categores = {}

for name in data['Category'].unique():
    m=0
    for i in data['Category']:
        if(i == name):
            m += 1
    categores[name] = m
categores
```

# Total Apps in a Type and percent

```python
f=0
for i in data['Type']:
    if(i == "Free"):
        f += 1
print("There are total",f,"free application")
p=0
for i in data['Type']:
    if(i == "Paid"):
        p += 1
print("There are total",p,"paid application")
```

```python
print(int(f*100/(f+p)),"%","application are free")
```
